# Plotting functions and derivatives

In this notebook we are going to explore what derivatives are how they relate to a function. To plot our example functions and derivatives we'll make use of numpy, scipy and matplotlib.

```python
import matplotlib.pyplot as plt
from scipy.misc import derivative
import numpy as np
```

In this example, we will give the function $f(x)=x^3$ as input, then calculate the derivative and plot both the function and its derivative. Remember powers are "**" in python and not ^!

```python
# defining the function
def function(x):
    return x**3
  
# calculating its derivative
def deriv(x):
    return derivative(function, x)
  
# defininf x-axis intervals
y = np.linspace(-6, 6)
  
# plotting the function
plt.plot(y, function(y), color='purple', label='Function')
  
# plotting its derivative
plt.plot(y, deriv(y), color='green', label='Derivative')
  
# formatting
plt.legend(loc='upper left')
plt.grid(True)
```

Differential equations will have multiple solutions. We can see this most easily but looking at it inversly, i.e. by seeing that functions of the same form have the same derivative. First let's plot the function $x^2$ and see what its derivative looks like.

```python
# defining the function
def function(x):
    return x**2
  
# calculating its derivative
def deriv(x):
    return derivative(function, x)
  
# defininf x-axis intervals
y = np.linspace(-6, 6)
  
# plotting the function
plt.plot(y, function(y), color='purple', label='Function')
  
# plotting its derivative
plt.plot(y, deriv(y), color='green', label='Derivative')
  
# formatting
plt.legend(loc='upper left')
plt.grid(True)
```
