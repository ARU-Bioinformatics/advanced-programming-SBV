# Getting ODEs from SMBL format to Python 

Models in the Biomodels database are written in Systems Biology Markup Language ( [SBML](https://sbml.org/)). This is a standard for how systems biology models are presented and stored for the community, but is not very usable. We ideally want to run our models in Python, or at the very least get the models in python nomenclature. The SBMLtoODEpy project includes a set of example sbml_files for running demos in the repo you can clone. This tutorial uses these example files, and then demonstrates how to access any particular model from [Biomodels](https://www.ebi.ac.uk/biomodels/) at the end.

## Example 1: Borisov2009

```
sbmltoodepy.ParseAndCreateModel("Borisov2009_insulin_EGF.xml", outputFilePath = "PythonFile.py", className = "ModelName")
```

```
from PythonFile import ModelName
modelInstance = ModelName()
```

```
# get the dictionary keys for the IDs of the species in the model
print(modelInstance.s.keys())
```

```
# get the dictionary keys for the IDs of the compartments in the model
print(modelInstance.c.keys())
```

```
# get the dictionary keys for the IDs of the parameters in the model
print(modelInstance.p.keys())
```

```
# replace compartmentId with one of the dictionary keys returned from print(modelInstance.c.keys())
print(modelInstance.c['cell'].size) 
# replace parameterId with one of the dictionary keys returned from print(modelInstance.p.keys())
print(modelInstance.p['k1'].value)
# replace speciesId with one of the dictionary keys returned from print(modelInstance.s.keys())
print(modelInstance.s['EGF'].concentration)
print(modelInstance.s['EGF'].amount)
```

```
print(modelInstance.time)
```

```
timeinterval = 1
modelInstance.RunSimulation(timeinterval)
```

```
print(modelInstance.time)
```

```
import numpy as np
times = np.zeros(101)
times[0] = modelInstance.time
concentrations = np.zeros(101)
concentrations[0] = modelInstance.s['EGF'].concentration
timeinterval = 1
for i in range(100):
	modelInstance.RunSimulation(timeinterval)
	times[i+1] = modelInstance.time
	concentrations[i+1] = modelInstance.s['EGF'].concentration
```

```
import matplotlib.pyplot as plt
plt.plot(times,concentrations)
```

## Example 2: Guyton1972

```
sbmltoodepy.ParseAndCreateModel("Guyton1972_Angiotensin.xml",outputFilePath = "GuytonFile.py", className = "GuytonModel")
from GuytonFile import GuytonModel
model = GuytonModel()
```

```
# get the dictionary keys for the IDs of the parameters in the model
print(model.p.keys())
model.RunSimulation(1)
```

## Example 3: Locke 2005

The above examples utilised SMBL files in the SMBLtoODEpy repo. We will want to utilise the SMBL for our model of choice, but how do we get them? We can simply find the download link from the Biomodels website, and then use wget to bring in our file. An example is shown below. However, there are a few things to note:

*   You need the model file that includes "urn" in the file name.
*   Check that your model is suitable for the ODE solver in the python script generated e.g. is it a linear model!

biomodels link.png

```
! wget https://www.ebi.ac.uk/biomodels/model/download/BIOMD0000000298.2?filename=BIOMD0000000298_urn.xml -O BIOMD0000000298_urn.xml
```

```
sbmltoodepy.ParseAndCreateModel("BIOMD0000000298_urn-L2V4.xml",outputFilePath = "Lockefile.py", className = "Lockemodel")
from Lockefile import Lockemodel
model = Lockemodel()
```

```
# get the dictionary keys for the IDs of the parameters in the model
print(model.p.keys())
print(model.s.keys())
print(model.c.keys())
```
